```python
# Filtrando dados com nome do time '2TM' ou '3TM'
data_filtered = data[~data['Tm'].isin(['2TM', '3TM'])]
data_filtered
```

```python
# Somando o total de jardas (YDS) por time
total_jardas_time = data_filtered.groupby('Tm')['Yds'].sum().sort_values(ascending=False)
total_jardas_time.head()
```

```python
# Plotando gráfico de barras com o total de jardas por time
plt.figure(figsize=(12, 6))
total_jardas_time.plot(kind='bar', color='red')
plt.title('Total de jardas passadas por time')
plt.xlabel('Time')
plt.ylabel('Total de jardas passadas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
media_jardas_time = data_filtered.groupby('Tm')['Yds'].mean().sort_values(ascending=False)
media_jardas_time.head()
```
